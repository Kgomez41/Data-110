```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

try:
    df = pd.read_csv('happiness2020.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'happiness2020.csv' not found.")
    df = None # Assign None to df in case of error
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

# New Section

```python
yellow_pastel = "#FFF7AE"
pink_pastel = "#FFB6C1"

top10 = df.nlargest(10, 'Ladder score')[['Country name', 'Ladder score']]

plt.figure(figsize=(10,6))
sns.barplot(data=top10, x='Ladder score', y='Country name', color=pink_pastel)
plt.title('Top 10 Happiest Countries (2020)', fontsize=14, color=yellow_pastel)
plt.xlabel('Ladder Score', color=yellow_pastel)
plt.ylabel('Country', color=yellow_pastel)
plt.gca().set_facecolor("#FFF8DC")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Regional indicator', y='Ladder score', color=yellow_pastel)
plt.title('Happiness Score Distribution by Region', fontsize=14, color=pink_pastel)
plt.xticks(rotation=90, color=pink_pastel)
plt.ylabel('Ladder Score', color=pink_pastel)
plt.xlabel('Region', color=pink_pastel)
plt.gca().set_facecolor("#FFFAF0")
plt.tight_layout()
plt.show()
```
