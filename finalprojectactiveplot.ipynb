```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("happiness2020.csv")

fig = px.scatter(
    df,
    x="Logged GDP per capita",
    y="Ladder score",
    color="Regional indicator",
    hover_name="Country name",
    size="Healthy life expectancy",
    title="💗 Happiness vs. GDP per Capita (2020)",
    labels={
        "Logged GDP per capita": "GDP per Capita (log)",
        "Ladder score": "Happiness Score",
        "Regional indicator": "Region"
    },
    color_discrete_sequence=px.colors.sequential.Pinkyl  # PINK theme!
)

fig.update_layout(
    title_x=0.5,
    paper_bgcolor="mistyrose",
    plot_bgcolor="lavenderblush",
    font=dict(family="Arial", size=14, color="deeppink")
)

fig.show()
```

```python
fig = px.choropleth(
    df,
    locations="Country name",
    locationmode="country names",
    color="Ladder score",
    hover_name="Country name",
    color_continuous_scale=px.colors.sequential.Pinkyl,  # pink color scale
    title="💗 Global Happiness Scores by Country (2020)",
    labels={"Ladder score": "Happiness Score"}
)

fig.update_layout(
    title_x=0.5,
    paper_bgcolor="mistyrose",
    font=dict(family="Arial", size=14, color="deeppink")
)

fig.show()
```
