```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset
crime_data = {
    "Offense": ["Aggravated Assault", "Burglary", "Larceny", "Motor Vehicle Theft", "Murder", "Rape (revised definition)", "Robbery"],
    "2023 Rate": [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5],
    "Change (%)": [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
}

df = pd.DataFrame(crime_data)
```

```python
# Task 1: Basic Bar Chart for Crime Rates in 2023
plt.figure(figsize=(10, 6))
sns.barplot(x="2023 Rate", y="Offense", data=df, palette="Reds_r")
plt.xlabel("Offenses per 100,000 People")
plt.ylabel("Offense Type")
plt.title("Crime Rates per 100,000 People in 2023")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Task 2: Horizontal Bar Chart for Percentage Change in Crime Rates
plt.figure(figsize=(10, 6))
colors = sns.color_palette("pink", len(df))
sns.barplot(x="Change (%)", y="Offense", data=df, palette=colors)
plt.axvline(x=0, color='purple', linewidth=1)
plt.xlabel("Percentage Change (2022-2023)")
plt.ylabel("Offense Type")
plt.title("Percentage Change in Crime Rates (2022-2023)")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Optional Task: Combined Bar Chart for 2022 and 2023 Crime Rates
df["2022 Rate"] = df["2023 Rate"] / (1 + df["Change (%)"] / 100)
plt.figure(figsize=(10, 6))
bar_width = 0.4
indices = np.arange(len(df))

plt.barh(indices - bar_width/2, df["2022 Rate"], bar_width, label="2022", color="pink")
plt.barh(indices + bar_width/2, df["2023 Rate"], bar_width, label="2023", color="purple")

plt.yticks(indices, df["Offense"])
plt.xlabel("Offenses per 100,000 People")
plt.ylabel("Offense Type")
plt.title("Comparison of Crime Rates (2022 vs 2023)")
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
