```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 1: Setup and Styles
print(plt.style.available)

# Use the updated seaborn style name
plt.style.use('seaborn-v0_8-whitegrid')  # Default style

alternative_style = 'ggplot'  # Chosen alternative style
```

```python
# Task 2: Basic Line Plot
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.figure()
plt.plot(x, y, label='$f(x) = x^2 - 2x + 1$')
plt.legend()
plt.title('Quadratic Function')
plt.show()
```

```python
# Task 3: Multiple Lines
x = np.linspace(-2, 2, 100)
y1 = x**3
y2 = np.exp(x)
y3 = np.log(x[x > 0])
plt.figure()
plt.plot(x, y1, label='$x^3$', color='k')
plt.plot(x, y2, label='$e^x$', color='b')
plt.plot(x[x > 0], y3, label='$\log(x)$', color='c')
plt.legend()
plt.title('Multiple Functions')
plt.show()
```

```python
# Task 4: Styling Lines
x = np.linspace(-2, 2, 100)
plt.figure()
plt.plot(x, x**2, 'k-', label='$x^2$')
plt.plot(x, x**3, 'g--', label='$x^3$')
plt.plot(x, x**4, 'c-.', label='$x^4$')
plt.plot(x, x**5, 'm:', label='$x^5$')
plt.legend()
plt.title('Different Line Styles')
plt.show()
```

```python
# Task 5: Axes Limits and Scaling
x = np.linspace(-10, 10, 400)
y = 1 / (x**2 + 1)
plt.figure()
plt.plot(x, y)
plt.xlim(-5, 5)
plt.ylim(0, 1)
plt.title('Custom Axes Limits')
plt.show()

plt.figure()
plt.plot(x, y)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title('Reversed Axes')
plt.show()

plt.figure()
plt.plot(x, y)
plt.axis('tight')
plt.title('Tight Axis')
plt.show()

plt.figure()
plt.plot(x, y)
plt.axis('equal')
plt.title('Equal Axis')
plt.show()
```

```python
# Task 6: Labels and Legends
x = np.linspace(0, 4, 100)
plt.figure()
plt.plot(x, x, label='$x$')
plt.plot(x, x**(1/3), label='$x^{1/3}$')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Labels and Legends')
plt.legend()
plt.show()
```

```python
# Task 7: Basic Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
plt.plot(x, y, 'o')
plt.title('Scatter Plot using plot()')
plt.show()

plt.figure()
plt.scatter(x, y)
plt.title('Scatter Plot using scatter()')
plt.show()
```

```python
# Task 8: Customizing Scatter Plots
x = np.linspace(0, 4*np.pi, 100)
y = x * np.sin(x)
plt.figure()
plt.scatter(x, y, s=50, c='red', edgecolors='black', linewidths=1.5)
plt.plot(x, y, 'b-', linewidth=2)
plt.title('Customized Scatter Plot')
plt.show()
```

```python
# Task 9: Combined Plot
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = x * np.cos(x)
x_scatter = np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y_scatter = x_scatter * np.cos(x_scatter)
plt.figure()
plt.plot(x, y, 'r-', label='$x \cos(x)$')
plt.scatter(x_scatter, y_scatter, c='r', label='Scatter Points')
plt.legend()
plt.title('Combined Line and Scatter Plot')
plt.show()
```

```python
# Task 6: Labels and Legends
x = np.linspace(0, 4, 100)
plt.figure()
plt.plot(x, x, label='$x$')  # Fixed: Closed the LaTeX expression with $
plt.plot(x, x**(1/3), label='$x^{1/3}$')  # Fixed: Closed the LaTeX expression with $
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Labels and Legends')
plt.legend()
plt.show()
```

```python
# Task 11: Style Comparison
styles = ['default', 'seaborn-whitegrid', alternative_style]
x = np.linspace(0, 4*np.pi, 100)
y = x * np.sin(x)
for style in styles:
    plt.style.use
    plt.figure()
    plt.plot(x, y, 'b-', linewidth=2)
    plt.scatter(x[::10], y[::10], s=50, c='red', edgecolors='black')
    plt.title(f'Style: {style}')
    plt.show()
```
